```python
# I import all the libraries that I need in this project.
# Each import has a specific job, and I explain them as I go.

import zipfile                     # lets me open and extract .zip files

import pandas as pd                # pandas is used to handle tables (dataframes)
import numpy as np                 # numpy is used for numerical operations and arrays

# From sklearn.model_selection I import tools related to splitting and evaluation.
from sklearn.model_selection import (
    train_test_split,              # splits the dataset into training and testing parts
    GridSearchCV,                  # tries many hyperparameter combinations and picks the best one
    learning_curve,                # helps me see how performance changes when I add more training data
    StratifiedKFold                # makes cross-validation folds that keep class balances the same
)

# From sklearn.preprocessing I import StandardScaler.
from sklearn.preprocessing import StandardScaler   # scales features so they are on a similar range

# Pipeline lets me chain preprocessing (scaling) and the model into one object.
from sklearn.pipeline import Pipeline              # connects scaler + model together

# These are metrics that I will use to judge how good the models are.
from sklearn.metrics import (
    accuracy_score,                 # percentage of correct predictions
    classification_report,          # gives precision, recall, and F1-score per class
    confusion_matrix,               # table that compares true vs predicted labels
    precision_recall_fscore_support # gives precision, recall, F1, and support in arrays
)

# Here I import the three different models that I want to compare.
from sklearn.ensemble import RandomForestClassifier   # tree-based ensemble model
from sklearn.svm import SVC                          # Support Vector Machine classifier
from sklearn.neural_network import MLPClassifier     # Multi-Layer Perceptron (neural network)

# These are plotting libraries so I can visualize results.
import matplotlib.pyplot as plt
import seaborn as sns

# This magic command tells Jupyter/Colab to show plots inside the notebook cells.
%matplotlib inline

print("✅ Imports done.")  # simple confirmation that imports worked correctly

# ------------------------------------------------------------------------------
# ICS350 - Assignment #3
# Team 3
# Hessa AlNaqbi (202306035)
# Manal AlMarzooqi (202320606)
# Amna Ibrahim (202318538)
# Almaha Ibrahim (202230136)
# GTZAN Music Genre Classification – Part 3: Experiments and Results
#
# In this notebook, I do the following:
# 1. Load the GTZAN 30-second feature dataset from a CSV file.
# 2. Separate the data into input features (X) and labels (y).
# 3. Split the data into training and test sets using a stratified split.
# 4. Build three models: Random Forest, SVM, and MLP, each inside a Pipeline.
# 5. Use GridSearchCV to tune hyperparameters using 5-fold cross-validation.
# 6. Evaluate the best models on the test set using accuracy and macro F1.
# 7. Show per-genre metrics, a confusion matrix, feature importances, and a learning curve.
# ------------------------------------------------------------------------------

# --- UNZIP THE DATASET FILE ---

# The dataset is stored in a compressed .zip file called 'features_30_sec.csv.zip'.
# In Colab I upload this file in the Files panel. Here I open it and extract the CSV.
with zipfile.ZipFile("features_30_sec.csv.zip", 'r') as zip_ref:
    zip_ref.extractall()   # this creates 'features_30_sec.csv' in the current working directory

# --- LOAD THE GTZAN FEATURE DATASET ---

# Now I read the extracted CSV file into a pandas DataFrame.
# 'df' will contain 1000 rows, each row is one music clip, and many columns of features.
df = pd.read_csv('features_30_sec.csv')

print('Columns in the dataset:')
print(df.columns.tolist())   # I print the column names to understand what features I have.

print('\nFirst 5 rows:')
display(df.head())           # I look at the first 5 rows to get a feeling of the data values.

# --- DEFINE FEATURES AND TARGET VARIABLES ---

# In this dataset:
# - 'filename' is just the name of the audio file, which is not a numeric feature.
# - 'label' is the genre (blues, rock, etc.), which is what I want to predict.
# So I do NOT want to use these two columns as input features.
non_feature_cols = ['filename', 'label']

# All the other columns are numeric features (for example MFCCs, tempo, etc.).
# I build a list that contains the names of all feature columns.
feature_cols = [c for c in df.columns if c not in non_feature_cols]

# X holds the input features: a matrix of shape (number of songs, number of features).
# Each row is a song, each column is a numeric descriptor.
X = df[feature_cols]

# y holds the target labels: this is the genre of each song.
y = df['label']

# I print the shapes to confirm they match what I expect.
print('\nFeature matrix shape:', X.shape)      # expected (1000, 58)
print('Target vector shape:', y.shape)        # expected (1000,)
print('Unique genres:', sorted(y.unique()))   # I check the list of 10 genres (blues, classical, etc.)

# --- STRATIFIED TRAIN/TEST SPLIT ---

# Here I split the data into training and test sets.
# test_size=0.2 means 20% of the data (200 songs) will be used for testing,
# and the remaining 80% (800 songs) will be used for training.
# stratify=y means that the class distribution (proportion of each genre)
# is preserved in both train and test sets, so they are balanced.
# random_state=42 is just a fixed seed so the split is reproducible.
X_train, X_test, y_train, y_test = train_test_split(
    X, y,
    test_size=0.2,
    stratify=y,
    random_state=42
)

print('Training samples:', X_train.shape[0])   # should be 800 rows
print('Test samples:', X_test.shape[0])        # should be 200 rows

# --- DEFINE MODEL PIPELINES WITH STANDARDIZATION ---

# Many machine learning algorithms prefer features that have similar scales
# (for example, mean ~0 and standard deviation ~1). StandardScaler does this.
# I use Pipelines so that scaling and model fitting happen together.
# This also prevents data leakage because scaling is learned only on the training data.

# Random Forest pipeline:
rf_pipeline = Pipeline([
    # First step: StandardScaler scales all numeric features.
    ('scaler', StandardScaler()),
    # Second step: RandomForestClassifier is the actual model.
    ('rf', RandomForestClassifier(random_state=42))
])

# SVM pipeline:
svm_pipeline = Pipeline([
    ('scaler', StandardScaler()),  # SVM is very sensitive to feature scales
    ('svm', SVC())                 # SVC is the SVM classifier; hyperparameters will be tuned later
])

# MLP (neural network) pipeline:
mlp_pipeline = Pipeline([
    ('scaler', StandardScaler()),  # neural networks also benefit a lot from scaled inputs
    ('mlp', MLPClassifier(
        max_iter=500,              # limit the number of training iterations
        random_state=42            # fixed seed for reproducible results
    ))
])

# --- HYPERPARAMETER GRIDS FOR GRIDSEARCHCV ---

# Here I define the hyperparameter search spaces for each model.
# GridSearchCV will try all combinations inside these dictionaries.

# For Random Forest:
rf_param_grid = {
    'rf__n_estimators': [100, 200, 300],   # how many trees to build
    'rf__max_depth': [10, 20, 30, None],   # maximum depth of each tree (None = full depth)
    'rf__min_samples_split': [2, 5, 10],   # minimum number of samples required to split a node
    'rf__min_samples_leaf': [1, 2, 4]      # minimum number of samples at a leaf node
}

# For SVM:
svm_param_grid = {
    'svm__C': [0.1, 1, 10, 100],           # regularization strength; higher C fits training data more
    'svm__gamma': ['scale', 'auto', 0.001, 0.01, 0.1],  # kernel coefficient; controls decision boundary shape
    'svm__kernel': ['rbf', 'poly', 'sigmoid']           # which kernel function to use
}

# For MLP (neural network):
mlp_param_grid = {
    # Different hidden layer configurations: each tuple defines the size of each hidden layer.
    'mlp__hidden_layer_sizes': [
        (128,),                            # one hidden layer with 128 neurons
        (256, 128),                        # two hidden layers: 256 then 128 neurons
        (256, 128, 64)                     # three hidden layers: 256 → 128 → 64
    ],
    # Different starting learning rates for gradient descent.
    'mlp__learning_rate_init': [0.001, 0.01, 0.1],
    # Alpha controls L2 regularization (helps prevent overfitting).
    'mlp__alpha': [0.0001, 0.001, 0.01],
    # Different batch sizes (how many samples we use per weight update step).
    'mlp__batch_size': [32, 64, 128]
}

# --- GRID SEARCH FOR RANDOM FOREST, SVM, AND MLP ---

# Now I create one GridSearchCV object for each pipeline.
# All of them:
# - use scoring='f1_macro', which treats all genres equally regardless of how many samples they have.
# - use cv=5, meaning 5-fold cross-validation.
# - use n_jobs=-1, which tells sklearn to use all CPU cores to speed things up.
# - use verbose=1, so I can see progress messages in the output.

rf_grid = GridSearchCV(
    estimator=rf_pipeline,     # model pipeline (scaler + RF)
    param_grid=rf_param_grid,  # hyperparameter search space for RF
    scoring='f1_macro',        # macro F1 score as the evaluation metric
    cv=5,                      # 5-fold cross-validation
    n_jobs=-1,                 # use all processors
    verbose=1                  # print progress information
)

svm_grid = GridSearchCV(
    estimator=svm_pipeline,    # pipeline for SVM
    param_grid=svm_param_grid,
    scoring='f1_macro',
    cv=5,
    n_jobs=-1,
    verbose=1
)

mlp_grid = GridSearchCV(
    estimator=mlp_pipeline,    # pipeline for MLP
    param_grid=mlp_param_grid,
    scoring='f1_macro',
    cv=5,
    n_jobs=-1,
    verbose=1
)

# I now fit the three GridSearchCV objects on the training data.
# This is the most time-consuming part, especially for SVM and MLP,
# because many combinations and folds are being trained.

print('Fitting Random Forest...')
rf_grid.fit(X_train, y_train)                   # trains multiple RF models and picks the best one
print('Best RF params:', rf_grid.best_params_)  # prints the best hyperparameters found
print('Best RF CV macro-F1:', rf_grid.best_score_)  # shows the best cross-validation macro F1 score

print('\nFitting SVM...')
svm_grid.fit(X_train, y_train)
print('Best SVM params:', svm_grid.best_params_)
print('Best SVM CV macro-F1:', svm_grid.best_score_)

print('\nFitting MLP...')
mlp_grid.fit(X_train, y_train)
print('Best MLP params:', mlp_grid.best_params_)
print('Best MLP CV macro-F1:', mlp_grid.best_score_)

# --- EVALUATE ALL THREE BEST MODELS ON THE TEST SET ---

# After GridSearchCV, each object has a 'best_estimator_'.
# This is the pipeline (scaler + model) using the best hyperparameters.
rf_best = rf_grid.best_estimator_
svm_best = svm_grid.best_estimator_
mlp_best = mlp_grid.best_estimator_

# I store them in a dictionary with readable names so I can loop over them.
models = {
    'Random Forest': rf_best,
    'SVM (RBF/other best)': svm_best,
    'Neural Network (MLP)': mlp_best
}

# This list will hold the summary metrics (accuracy, macro precision, etc.) for each model.
results = []

# I evaluate each model on the independent test set (X_test, y_test).
for name, model in models.items():
    # Get predictions from the model.
    y_pred = model.predict(X_test)

    # Calculate accuracy.
    acc = accuracy_score(y_test, y_pred)

    # Calculate macro-averaged precision, recall, and F1-score.
    # average='macro' means we compute the metric for each class and then take the average.
    # zero_division=0 ensures that if a class has no predicted samples, we just put 0 instead of error.
    prec, rec, f1, _ = precision_recall_fscore_support(
        y_test, y_pred, average='macro', zero_division=0
    )

    print(f'\n{name}')
    print('Accuracy:', acc)
    # classification_report prints precision, recall, and F1-score for each individual genre.
    print(classification_report(y_test, y_pred, zero_division=0))

    # I save the results in a dictionary and append it to the list.
    results.append({
        'Model': name,
        'Accuracy': acc,
        'Macro Precision': prec,
        'Macro Recall': rec,
        'Macro F1-Score': f1
    })

# I convert the results list into a DataFrame.
# This looks like a nice table (similar to Table 3.1 in the report).
results_df = pd.DataFrame(results)
print('\nSummary results (Table 3.1 style):')
display(results_df)

# --- PER-GENRE PERFORMANCE FOR THE BEST MODEL (SVM) ---

# For more detailed analysis, I focus on the SVM model.
best_model_name = 'SVM (RBF/other best)'
best_model = svm_best

# Predict the test labels with the best SVM.
y_pred_best = best_model.predict(X_test)

# Get a sorted list of genres to keep the order consistent everywhere.
labels = sorted(y.unique())

# Here I compute precision, recall, F1-score, and support for each genre separately.
precisions, recalls, f1s, supports = precision_recall_fscore_support(
    y_test, y_pred_best, labels=labels, zero_division=0
)

# I put these per-genre results into a DataFrame.
# This will resemble Table 3.2 in the report.
per_class_df = pd.DataFrame({
    'Genre': labels,
    'Precision': precisions,
    'Recall': recalls,
    'F1-Score': f1s,
    'Support': supports
})

print('Per-genre metrics for best model (Table 3.2 style):')
display(per_class_df)

# --- CONFUSION MATRIX HEATMAP FOR THE BEST MODEL (SVM) ---

# A confusion matrix shows where the model gets things right and where it gets confused.
# Rows are true genres, columns are predicted genres.
cm = confusion_matrix(y_test, y_pred_best, labels=labels)

plt.figure(figsize=(10, 8))   # make the figure large so labels are readable

# sns.heatmap draws a colored grid where each cell's color depends on the count.
sns.heatmap(
    cm,
    annot=True,                # write the numbers in each cell
    fmt='d',                   # 'd' means integer format
    xticklabels=labels,        # put genre names on the x-axis
    yticklabels=labels         # put genre names on the y-axis
)

plt.title('Confusion Matrix – Best Model')
plt.xlabel('Predicted Label')
plt.ylabel('True Label')
plt.tight_layout()             # adjust spacing so everything fits
plt.show()

# --- FEATURE IMPORTANCE FOR RANDOM FOREST (TOP 15 FEATURES) ---

# Random Forest gives us a feature_importances_ attribute.
# It tells us how important each feature was when the trees made decisions.
rf_model = rf_best.named_steps['rf']    # I get the RF model inside the pipeline.

importances = rf_model.feature_importances_   # array of importance values

feature_names = feature_cols                  # the names of the features

# argsort returns indices that would sort the array.
# [::-1] reverses the order so that we go from most important to least.
indices = np.argsort(importances)[::-1]

top_n = 15    # I only want to show the top 15 most important features.

plt.figure(figsize=(10, 6))

# I plot a bar chart for the top 15 feature importances.
plt.bar(range(top_n), importances[indices[:top_n]])

# I set the x-axis tick labels to the actual feature names,
# and rotate them so they don’t overlap too much.
plt.xticks(
    range(top_n),
    [feature_names[i] for i in indices[:top_n]],
    rotation=45,
    ha='right'
)

plt.title('Top 15 Feature Importances (Random Forest)')
plt.tight_layout()
plt.show()

# --- LEARNING CURVE FOR THE BEST SVM MODEL ---

# I create a StratifiedKFold object for cross-validation that keeps genres balanced in each fold.
cv = StratifiedKFold(
    n_splits=5,       # split into 5 folds
    shuffle=True,     # shuffle before splitting
    random_state=42   # fixed seed for reproducibility
)

# learning_curve will train the model on different fractions of the data
# and compute training and validation scores for each fraction.
train_sizes, train_scores, val_scores = learning_curve(
    svm_best,                    # the best SVM pipeline
    X, y,                        # full dataset (features and labels)
    cv=cv,                       # stratified 5-fold CV
    train_sizes=np.linspace(0.1, 1.0, 5),  # uses 10%, 32.5%, 55%, 77.5%, and 100% of data
    scoring='f1_macro',          # again, macro F1 score
    n_jobs=-1                    # run folds in parallel
)

# train_scores and val_scores are 2D arrays: (number of train_sizes, number of folds).
# I take the mean and standard deviation across folds for each train size.
train_mean = train_scores.mean(axis=1)
train_std = train_scores.std(axis=1)

val_mean = val_scores.mean(axis=1)
val_std = val_scores.std(axis=1)

plt.figure(figsize=(8, 6))

# Plot training performance curve.
plt.plot(train_sizes, train_mean, marker='o', label='Training F1 (macro)')
plt.fill_between(
    train_sizes,
    train_mean - train_std,
    train_mean + train_std,
    alpha=0.2
)

# Plot validation performance curve.
plt.plot(train_sizes, val_mean, marker='s', label='Validation F1 (macro)')
plt.fill_between(
    train_sizes,
    val_mean - val_std,
    val_mean + val_std,
    alpha=0.2
)

plt.title('Learning Curve – SVM (Best Model)')
plt.xlabel('Number of Training Samples')   # x-axis is how many training samples we used
plt.ylabel('Macro F1-Score')               # y-axis is the macro F1-score
plt.legend()
plt.grid(True)
plt.tight_layout()
plt.show()
```
